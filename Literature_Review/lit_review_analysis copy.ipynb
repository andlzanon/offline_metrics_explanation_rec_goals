```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lit_review = pd.read_csv("lit_review.csv")
lit_review
```

```python
lit_review2 = pd.read_csv("lit_review_2.csv")
lit_review2
```

```python
lit_review = pd.concat([lit_review,lit_review2])
lit_review
```

```python
lit_review["Booktitle"] = lit_review["Booktitle"].str.replace(r'\b\d{1,2}(st|nd|rd|th)\b', '', regex=True) \
                                           .str.replace(r'\b(19|20)\d{2}\b', '', regex=True) \
                                           .str.strip()
lit_review["Booktitle"] = lit_review["Booktitle"].fillna("").str.replace(r' +', ' ', regex=True)
lit_review["Journal"] = lit_review["Journal"].fillna("").str.replace(r' +', ' ', regex=True)
lit_review
```

```python
lit_review["Journal_Book"]= lit_review["Booktitle"].astype(str) + lit_review["Journal"].astype(str)
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('BigData', 'Big Data')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Eleventh ', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('on Conference ', 'Conference ')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('''\\&amp;''', 'and')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('SIGKDD International', 'SIGKDD')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Annual', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace(' in the Asia Pacific Region', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Proceedings of the World Wide Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('The World Wide Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Proceedings of the Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('The Web Conference', 'Proceedings of the ACM Web Conference')
```

```python
lit_review
```

```python
count_df = lit_review.groupby("Journal_Book").count()
count_df = count_df[count_df.columns[:1]]
count_df.columns = ["count"]
count_df = count_df.sort_values("count", ascending=False)
print(count_df.columns)
count_df
```

```python
lit_review.groupby("Year").count()
```

```python
plt.figure(figsize=(10, 12))
plt.plot(count_df['count'], np.arange(len(count_df)), 'o', markersize=8)
plt.yticks(np.arange(len(count_df)), count_df.index, fontsize=10)
plt.xlabel('Count')
plt.title('Conference/Journal Counts')
plt.show()
```

```python
ax = sns.histplot(lit_review["Year"],  discrete=True, shrink=.8)
ax.bar_label(ax.containers[0], label_type='edge')
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Frequency by Year")
plt.xticks(ticks=lit_review["Year"].unique())
plt.show()
```
