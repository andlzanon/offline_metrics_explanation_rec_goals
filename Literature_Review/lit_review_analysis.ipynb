```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lit_review = pd.read_csv("lit_review.csv")
lit_review
```

```python
lit_review2 = pd.read_csv("lit_review_2.csv")
lit_review2
```

```python
lit_review = pd.concat([lit_review,lit_review2])
lit_review
```

```python
lit_review["Booktitle"] = lit_review["Booktitle"].str.replace(r'\b\d{1,2}(st|nd|rd|th)\b', '', regex=True) \
                                           .str.replace(r'\b(19|20)\d{2}\b', '', regex=True) \
                                           .str.strip()
lit_review["Booktitle"] = lit_review["Booktitle"].fillna("").str.replace(r' +', ' ', regex=True)
lit_review["Journal"] = lit_review["Journal"].fillna("").str.replace(r' +', ' ', regex=True)
lit_review
```

```python
import re 
lit_review["Journal_Book"]= lit_review["Booktitle"].astype(str) + lit_review["Journal"].astype(str)
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('BigData', 'Big Data')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Eleventh ', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('on Conference ', 'Conference ')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('''\\&amp;''', 'and')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('SIGKDD International', 'SIGKDD')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Annual', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace(' in the Asia Pacific Region', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Proceedings of the World Wide Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('The World Wide Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Proceedings of the Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('The Web Conference', 'Proceedings of the ACM Web Conference')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('Proceedings of the ', '')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('ACM SIGKDD Conference on Knowledge Discovery and Data Mining', 'ACM SIGKDD Conference on Knowledge Discovery \\& Data Mining')
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('ACM International Conference on Information and Knowledge Management', 'ACM International Conference on Information \\& Knowledge Management')
lit_review["Journal_Book"] = lit_review["Journal_Book"].apply(lambda x: re.sub(r"\s+", " ", x).strip())
lit_review["Journal_Book"] = lit_review["Journal_Book"].str.replace('\\&', 'and')
```

```python
lit_review = lit_review.drop_duplicates(subset=['BibliographyType', 'ISBN', 'Identifier', 'Author', 'Title', 'Journal',
       'Volume', 'Number', 'Month', 'Pages', 'Year', 'Address', 'Note', 'URL',
       'Booktitle', 'Chapter', 'Edition', 'Series', 'Editor', 'Publisher',
       'ReportType', 'Howpublished', 'Institution', 'Organizations', 'School',
       'Annote', 'Journal_Book'])
lit_review
```

### Conclusions

- There are no consensus for explanation metrics in Recommender Systems - OK
- Very few papers do online expriments - OK 
- Limitation of NLP metrics such as BLEU and ROUGE. Offline metrics are vastly used but their impact on user experience is not evaluated - OK
- The vast majority of papers rely on other contributions evaluating explanations only as proof of concepts. Which lead to reproducibility issues  - OK
- Difference between user explanation and system explanation 
- Lack of evaluation protocol - OK
    - Difficult to evaluate under certain conditions such as on online evaluation - Future Works
- Increase of the study of multimodal and personality explanations - OK 

- Does explanation goals cover all the objetives of showing explanations to users?
- In cases where papers cite that with external metadata the recommendation becomes more transparent but do not actually explain how to generate explanations were considered intrinsic
- Different purposes of explanations (choose best review, explain an item, etc)
- Purposes of explanation: for a developer or a user, kinds of user
- Much theoretical work is needed
